## The σ-face Framework

From the Ainulindale conjecture, the critical line σ = ½ is the boundary between:
- **Below** (ZD, zero divisors): topological features, shapes, interference patterns — no stuff
- **Above** (CD, co-divisors): particles, plasma, stuff that radiates

The Faraday rotation measure φ (rad/m²) integrates B·n_e along the line of sight:

$$\phi = 0.812 \int_{\text{source}}^{\text{observer}} n_e \, \mathbf{B} \cdot d\mathbf{l} \quad [\text{rad/m}^2]$$

where n_e is electron density (cm⁻³) and B is magnetic field (μG), dl in pc.

**If DM is a ZD boundary** (no plasma, no electrons): the integral through the DM halo contributes nothing. φ is determined entirely by the ICM gas, which is displaced from the DM peaks in the Bullet Cluster.

**If DM is stuff** (has electrons, however few): the integral picks up a signal at the DM peak positions. The sign and amplitude encode B_DM · n_e_DM · depth_DM.

## The Measurement

Define a transect **perpendicular to the merger axis**, crossing through the DM band center
(midpoint between DM_NW and DM_SE peaks).

At each position along the transect, measure:
1. φ(l) — Faraday depth from MeerKAT RM synthesis [rad/m²]
2. κ(l) — gravitational convergence [dimensionless]
3. Σ_X(l) — X-ray surface brightness [counts/s/pixel]
4. y(l) — Compton-y parameter [dimensionless]

**The diagnostic quantity** (raw, no normalisation):

$$\Delta\phi_\kappa = \phi(l_{\kappa,\text{peak}}) - \hat{\phi}_{\text{gas}}(l_{\kappa,\text{peak}})$$

where $\hat{\phi}_{\text{gas}}$ is the RM expected from the gas alone, estimated by linear
interpolation between positions flanking the κ peak (where gas dominates).

No background subtraction. No sigma-clipping. No fitting.
