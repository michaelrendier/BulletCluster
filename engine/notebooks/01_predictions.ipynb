## Physical reasoning for each prediction

### P-RM (Faraday screen test)

Faraday depth φ (rad/m²) integrates n_e · B along every line of sight:

$$\phi = 0.812 \int n_e B_{\parallel} \, dl$$

**Wave DM**: n_e_DM = 0 by definition. The DM halo is a ZD boundary — a shape,
not plasma. φ is determined entirely by the ICM gas, which in the Bullet Cluster
is offset 2–4' from the κ peaks. Therefore φ is *low* at the κ peak positions
and *high* at the gas peaks. ΔRM at κ = 0.

**Particle DM**: DM halos must have some n_e if they contain any charged species.
Even if this n_e is tiny (1e-5 cm⁻³) the integrated column can be substantial
at Mpc depth. Any non-zero B in the DM halo (sourced by structure formation or
astrophysical seeding) creates a Faraday screen at the κ peak. ΔRM ≠ 0.

### P-PI (Polarised fraction at DM band center)

**Wave DM (slingshot/lensing)**: The DM gravitational potential acts as a lens.
Lines of sight passing through the potential well are magnified. Background
polarised sources and the radio relic synchrotron emission are magnified at
the DM positions → higher apparent polarised intensity.

**Particle DM**: A DM plasma creates internal Faraday dispersion. Different
lines of sight within the beam sample different RM values → beam depolarisation.
The net polarised fraction drops below the gas-only background level → dark band.

### P-EVPA (E-vector rotation)

EVPA(λ²) = EVPA₀ + RM·λ². Smooth RM = smooth EVPA gradient.
An abrupt RM jump at the κ peak = abrupt EVPA rotation crossing the DM boundary.
The gradient |dEVPA/dl| at the κ peak location is the measurable signature.

### P-CORR (RM–κ vs RM–Σ_X correlation)

If RM is produced by the gas alone: RM should correlate tightly with gas density
(n_e ∝ Σ_X^{1/2}). The RM–κ correlation should be low because κ peaks are offset
from gas peaks.

If DM has a Faraday contribution: RM will correlate with κ *in addition to*
Σ_X. Multiple regression R²(RM ~ Σ_X + κ) > R²(RM ~ Σ_X alone).
